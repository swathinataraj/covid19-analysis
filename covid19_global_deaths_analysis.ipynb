```python
covid1 = pd.read_csv('time_series_covid19_deaths_global.csv', low_memory=False)
```

```python
covid2 = covid1.drop(['Province/State','Lat','Long'], axis = 1) 
```

```python
covid2
```

```python
covid3 = pd.melt(covid2, id_vars=['Country/Region'], var_name="Date", value_name="Deaths")
```

```python
covid3['Date'] = pd.to_datetime(covid3['Date'], format="%m/%d/%y")
```

```python
covid3
```

```python
%who DataFrame
```

```python
covid3 = covid3.sort_values(['Country/Region', 'Date']).reset_index(drop = True)
```

```python
temp1 = covid3.groupby(['Country/Region','Date'])
```

```python
temp1
```

```python
#covid2.groupby('Province/State').sum()[['Deaths']]
covid3['Deaths'] = temp1.Deaths.transform('sum')
```

```python
covid3['Deaths'] = covid3['Deaths'].fillna(0).astype(int)
```

```python
#covid3.to_csv("res1.csv")
```

```python
covid3 = covid3.drop_duplicates(['Country/Region','Date'], keep='first')
```

```python
#covid3.to_csv("res3.csv")
```

```python
import sqlite3
conn = sqlite3.connect("covid19.db")
cur = conn.cursor()
covid3.to_sql("deaths_total", conn, if_exists="replace",index=False)
pd.read_sql_query("select * from deaths_total;", conn)
```

```python
grouped = covid3.groupby(['Country/Region'])
```

```python
list1 = []
for name, group in grouped:
    #print(len(group.index))
    new = [group.Deaths.values[0]]
    for i in range(1, len(group.index)):
        new.append(group.Deaths.values[i]-group.Deaths.values[i-1])
    list1.extend(new)
```

```python
covid3['DailyChange'] = list1
```
